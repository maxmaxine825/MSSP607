```python
df = pd.read_csv("data/Student Performance Data.csv")

df.head()
```

This lambda function takes the Student Performance Data Set as input and returns a new DataFrame that includes an additional column called “average_score.”

```python
add_avg = lambda df: df.assign(
    average_score = df[["math score", "reading score", "writing score"]].mean(axis=1).round(2)
)
```

```python
new_df = add_avg(df)

new_df.head()
```
